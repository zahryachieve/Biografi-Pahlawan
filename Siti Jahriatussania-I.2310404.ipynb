```python
import numpy as np
import matplotlib.pyplot as plt

# perbaikan: np.linspace (lowercase) dan koma antara argumen
x = np.linspace(-18, 18, 188)
y = np.tan(x)

# sembunyikan nilai ekstrim supaya tidak muncul spike di plot
y_masked = np.ma.masked_where(np.abs(y) > 20, y)

plt.figure(figsize=(8,4))
plt.plot(x, y_masked, label='tan(x)')
plt.ylim(-10, 10)   # batasi sumbu y supaya asimtot tidak mengacaukan tampilan
plt.grid(True)
plt.legend()
plt.show()

```
